```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv")
```

In this case, I decided to compare the #1 and #2 races with the most cases of diabetes.

```python
df
```

Groupby function showed which were the two groups with the highest number of cases as shown below.

```python
df.groupby(['race'])['encounter_id'].count()
```

I created a graph showing all races represented, in ascending order below.

```python
df.groupby(['race'])['encounter_id'].count().sort_values(ascending=True).plot(kind="bar")
```

African Americans and Caucasians fit my criteria so I filtered the two

```python
df2 = df[df['race'] == 'Caucasian']
df3 = df[df['race'] == 'AfricanAmerican']
```

```python
count_w_male = df2[df2['gender'] == 'Male'].shape[0]
count_w_female = df2[df2['gender'] == 'Female'].shape[0]
count_b_male = df3[df3['gender'] == 'Male'].shape[0]
count_b_female = df3[df3['gender'] == 'Female'].shape[0]
White_Males = ("Amount Of White Males With Diabetes: "+ str(count_w_male))
White_Females = ("Amount Of White Females With Diabetes: " + str(count_w_female))
Black_Males = ("Amount Of Black Males With Diabetes: "+ str(count_b_male))
Black_Females= ("Amount Of Black Females With Diabetes: "+ str(count_b_female))
print(White_Males)
print(White_Females)
print(Black_Males)
print(Black_Females)
```

Above I established the amount of males and females both Caucasian and African American and represented them Below in a comparison graph.

```python
labels = ['White Males', 'White Females', 'Black Males', 'Black Females']
counts = [count_w_male, count_w_female, count_b_male, count_b_female]
```

```python
plt.bar(labels, counts, color=['blue', 'pink', 'green', 'purple'])
plt.title('Diabetes Race and Gender Comparison')
plt.ylabel('Number of Cases')
plt.show()
```

```python
print(White_Males)
print(White_Females)
print(Black_Males)
print(Black_Females)
```

In conclusion to looking over this data I found the following: Caucasians had the highest total of diabetes cases with 76,099 while African Americans were second with a total of 19,210. Though the total number of cases between races  are not comparable the fact that women, both black and white, have the most cases in their race is interesting. Caucasian women have 3,279 cases more than their counterparts and African American women 4,246 cases more than theirs.
